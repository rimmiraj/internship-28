# 9) Write a python program to scrape mentioned details from dineout.co.in :

# import required libraies

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find('div',class_="restnt-info cursor")
first_title
```

```python
first_title.text
```

```python
loc = soup.find('div',class_="restnt-loc ellipsis")
loc
```

```python
loc.text
```

```python
price = soup.find('span',class_="double-line-ellipsis")
price
```

```python
price.text.split()[1]
```

```python
titles=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
    
    
    
titles    
```

```python
location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
    
    
    
location    
```

```python
price = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
    
    
    
price    
```

# making dataframe

```python
import pandas as pd
data = pd.DataFrame({'titles':titles,'location':location,'price':price})
```

```python
print(len(titles),len(location),len(price))
```

```python
data
```

```python
data.head(2)
```
