Ques. 5 Scrape the details of trending repositories on Github.com.
Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

```python
url4 = 'https://github.com/'       #Giving the URL
driver.get(url4)

```

```python
#Clicking on explore button then on trending, So that I can lend to the required page, From where I have to extract the data.

explore = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[1]/a')
explore.click()

trending = driver.find_element_by_xpath('/html/body/div[4]/main/div[1]/nav/div/a[3]')
trending.click()
```

```python
title = []                # Giving the empty list
description = []
contributer = []
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')         #Used the concept of BeautifulSoup, and extracted the data.
t = soup.find_all('h1',attrs={'h3 lh-condensed'})
d = soup.find_all('article',attrs={'Box-row'})
c = soup.find_all('div',attrs={'f6 color-fg-muted mt-2'})
```

```python
for i in t:                                                 #used for loop, So that all the data what I have extracted in, It can be converted into text,as I have done below.
    try:                                                    #used try and except method. So, that where No data is present, I can mention, "No details available"
        title.append(i.find('span').text.strip('\n'))       #Appending the text data
    except AttributeError:
        title.append("No details available")
for j in d:
    try:
        description.append(j.find('p').text.split('\n')[-2])
    except AttributeError:
        description.append("No details available")
for k in c:
    try:
        contributer.append(k.find('a').text)
    except AttributeError:
        contributer.append("No details available")

print(len(title))                                      #Printing the length 
print(len(description))
print(len(contributer))
```

```python
urls = []                 #Giving the empty list for the URL's

links = driver.find_elements_by_xpath('/html/body/div[4]/main/div[1]/nav/div/a[3]')         #Giving the Xpath for each repositories

for url in links:
    urls.append(url.get_attribute('href'))          #Extracting the url's for the repositories.              
urls
```

```python
repository_title = []                    #Giving the Empty List
repository_description = []
contributors_count = []
language_used = []

for i in urls:
    driver.get(i)                        #Iterating into each URl's and extracting the required data.
    time.sleep(3)
    try:
        title = driver.find_element_by_xpath('//h1[@class=" d-flex flex-wrap flex-items-center wb-break-word f3 text-normal"]/span[1]')
        repository_title.append(title.text.replace('\n',''))             #Appending the text data and Replacing the unrequired data
    except NoSuchElementException:
        repository_title.append('-')                         #Used try and except. So, that in case, I data is not there then also our code get's executed without giving error.
        
    time.sleep(1)
    try:
        descrip = driver.find_element_by_xpath('//div[@class="BorderGrid-cell"]//p')
        repository_description.append(descrip.text.replace('\n',''))
    except NoSuchElementException:
        repository_description.append('-')

    time.sleep(1)
    try:
        count = driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributors_count.append(count.text)
        
    except NoSuchElementException:
        contributors_count.append('-')
    time.sleep(1)
    try:
        lang = driver.find_element_by_xpath('//li[@class="d-inline"]//a//span')
        language_used.append(lang.text.replace('\n',''))
    except NoSuchElementException:
        language_used.append('-')
```

```python
github = pd.DataFrame({})
github['repository_title'] = repository_title
github['repository_description'] = repository_description
github['contributors_count']  = contributors_count
github['language_used'] = language_used
github
```
